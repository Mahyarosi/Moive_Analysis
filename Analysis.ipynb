```python
numeric_cols = ['BudgetUSD', 'US_BoxOfficeUSD', 'Global_BoxOfficeUSD', 'Opening_Day_SalesUSD', 'One_Week_SalesUSD', 'IMDbRating',
                 'RottenTomatoesScore', 'NumVotesIMDb', 'NumVotesRT']
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
for col in numeric_cols:
    sns.histplot(df[col], bins=50, kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
df.columns
```

```python
categorical_cols = [
    'Genre', 'Country', 'Director', 'LeadActor'
]
```

```python
for col in categorical_cols:
    df[col].value_counts().head(10).plot(kind='bar')
    plt.title(f'Top 10 most common {col}')
    plt.show()
```

```python
sns.scatterplot(data=df, x='BudgetUSD', y='Global_BoxOfficeUSD', alpha=0.3)
plt.title('Budget vs Global Box Office')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
sns.scatterplot(data=df, x='IMDbRating' , y='RottenTomatoesScore', alpha=0.3)
plt.title('IMDb Rating vs Rotten Tomatoes Score')
plt.show()
```

```python
df.columns
```

```python
genre_year_stats = df.groupby(['Genre', 'ReleaseYear']).agg({
    'Global_BoxOfficeUSD': 'mean',
    'BudgetUSD': 'mean',
    'IMDbRating': 'mean',
    'RottenTomatoesScore': 'mean'
}).reset_index()
```

```python
genre_year_stats
```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(data=genre_year_stats, x='ReleaseYear', y='Global_BoxOfficeUSD', hue='Genre')
plt.title('Average Global Box Office by Genre Over Years')
plt.show()
```

```python
top_actors = df.groupby('LeadActor')['IMDbRating'].mean().sort_values(ascending=False).head(24)
```

```python
top_actors
```

```python
top_actors.plot(kind='bar')
plt.title('Top 10 Lead Actors by Average IMDb Rating')
plt.show()
```
